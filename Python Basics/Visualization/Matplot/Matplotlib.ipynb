```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(0, 50, 1)
```

```python
plt.tight_layout()

plt.subplot(1,2,1)
plt.plot(a, -a**2, color='red')

plt.subplot(2,2,2)
plt.plot(a, a**2, 'green')
plt.tight_layout()

plt.subplot(4,2,6)
plt.plot(a, a**3, 'purple')
plt.tight_layout()

plt.subplot(4,2,8)
plt.plot(a, a**2, 'blue')
plt.tight_layout()

# plt.subplot(4,2,3)
# plt.plot(a, a**8)

plt.show()
```

```python
graf = plt.figure(figsize=(10, 10))

graf.add_subplot(1,2,1)
plt.plot(a, -a**2, color='red')

graf.add_subplot(2,2,2)
plt.plot(a, a**2, 'green')
plt.tight_layout()

graf.add_subplot(4,2,6)
plt.plot(a, a**3, 'purple')


graf.add_subplot(4,2,8)
plt.plot(a, a**2, 'blue')
```
