```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('Fluxo de Caixa.csv')
```

```python
df = pd.DataFrame(file)
df.head()
```

```python
df.iloc[:, 0:7]
```

```python
df.fillna(0, inplace=True)
```

```python
df_money = df.iloc[:, 0:7].set_index('Mês')
df_money.index.names = [None]
df_money[['Gastos', 'Aulas Individuais']] = df_money[['Gastos', 'Aulas Individuais']].astype(float)
df_money.info()
```

```python
for i in df_money.columns[3:]:
    df_money[f'{i} Acumulado'] = df_money[i].cumsum()
```

```python
df_money
```

```python
plt.plot(df_money[ 'Lucro Liquído'], color = 'green')

plt.title('Faturamento Liquido 2022', fontsize = 14)
plt.xlabel('Meses', fontsize = 12)
plt.xticks(rotation=90)
plt.ylabel('Valor $', fontsize = 12)

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(df_money['Lucro Bruto'])
plt.xticks(rotation = 90)
plt.title('Faturamento Bruto 2022')

plt.subplot(1, 2, 2)
plt.plot(df_money['Lucro Liquído'], color = 'green')
plt.xticks(rotation = 90)
plt.title('Faturamento Liquído 2022')

plt.tight_layout()
plt.show()
```
