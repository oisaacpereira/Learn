```python
# Importacao das bibliotecas para manipulacao do Dataframe

import pandas as pd
```

```python
# leitura do arquivo

file = pd.read_csv('top_movies.csv') 
```

```python
# Transformacao e visualizacao previa do df

df = pd.DataFrame(file)
df.head(14)
```

```python
# Visao geral dos dados do arquivo

df.info()
```

```python
#visualizacao das colunas que possuem dados nan

df.isnull().sum()
```

```python
# Importacao das colunas que serao utilizadas na manipulacao, exclusao de valores nulos e atribuicao a um sub df filtrado

df_ft = df[['year','title', 'box_office_(gross_usa)', 'genre', 'people_score']].dropna()
```

```python
# Sub data frame criado

df_ft.head(20)
```

```python
# Slicing da coluna box_office separando o cifrao e multiplicadores dos numeros e transformando em float

df_ft['box_office_(gross_usa)'] = df_ft['box_office_(gross_usa)'].apply(lambda x: str(x).replace('$', ''))
```

```python
valor = df_ft['box_office_(gross_usa)'].apply(lambda x: str(x).strip(x[-1]) if 'M' in x 
                                              else (str(x).strip(x[-1]) if 'K' in x else x)).astype(float)
```

```python
multiplicador = df_ft['box_office_(gross_usa)'].apply(lambda x: 1000000 if str(x)[-1] == 'M' 
                                                      else (1000 if str(x)[-1] == 'K' else x)).astype(float)
```

```python
# Multiplicando valor pelo multiplicador que estava em string e 
# criando nova coluna com valor final da bilheteria

df_ft['Bilheteria EUA $'] = multiplicador * valor
df_ft
```

```python
# Exclusao da coluna fatiada e criacao df manipulavel

df_clean = df_ft.drop('box_office_(gross_usa)', axis=1).drop_duplicates()
```

```python
#colunas data frame renomeado para PT-BR

df_clean.rename(columns={'year': 'Ano',
                        'title': 'Filme',
                        'genre': 'Gênero',
                        'people_score': 'Avaliação Popular'            
                        }, inplace=True)
```

```python
# Top 10 bilheterias ate 2019

df_clean.sort_values('Bilheteria EUA $', ascending=False).head(10)
```
